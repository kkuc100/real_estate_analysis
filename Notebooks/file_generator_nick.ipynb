```python
# get CC_realtor_merged_cleaned_adjusted.csv
CC_realtor_merged_cleaned_adjusted = pd.read_csv('../Data/CC_realtor_merged_cleaned_adjusted.csv')

# head
CC_realtor_merged_cleaned_adjusted.head()
```

```python
# get the number of nulls in the dataframe
print(CC_realtor_merged_cleaned_adjusted.isnull().sum())

# get the rows with nulls
print(CC_realtor_merged_cleaned_adjusted[CC_realtor_merged_cleaned_adjusted.isnull().any(axis=1)])

# hard coded zips that are not in the pgeocode database
zip_lat_lon = {
    2625: (41.6485, -70.3475),
    2031: (42.1579, -71.1762),
    2068: (42.0248, -71.1022),
    2425: (42.3318, -71.1212),
    2469: (42.2965, -71.2921),
    2154: (42.3765, -71.2356)
}

# 2601.0
# Name: zip_cluster, dtype: float64
# 2397    2062.0
# Name: zip_cluster, dtype: float64
# 5451    2375.0
# Name: zip_cluster, dtype: float64
# 7308    2446.0
# Name: zip_cluster, dtype: float64
# 8237    2482.0
# Name: zip_cluster, dtype: float64
# 8364    2453.0

# add zip clusters based on the above
CC_realtor_merged_cleaned_adjusted.loc[CC_realtor_merged_cleaned_adjusted['ZIP'] == 2625, 'zip_cluster'] = 2601.0
CC_realtor_merged_cleaned_adjusted.loc[CC_realtor_merged_cleaned_adjusted['ZIP'] == 2031, 'zip_cluster'] = 2062.0
CC_realtor_merged_cleaned_adjusted.loc[CC_realtor_merged_cleaned_adjusted['ZIP'] == 2068, 'zip_cluster'] = 2375.0
CC_realtor_merged_cleaned_adjusted.loc[CC_realtor_merged_cleaned_adjusted['ZIP'] == 2425, 'zip_cluster'] = 2446.0
CC_realtor_merged_cleaned_adjusted.loc[CC_realtor_merged_cleaned_adjusted['ZIP'] == 2469, 'zip_cluster'] = 2482.0
CC_realtor_merged_cleaned_adjusted.loc[CC_realtor_merged_cleaned_adjusted['ZIP'] == 2154, 'zip_cluster'] = 2453.0

# add the lat and long for the hard coded zips
for zip_code, (lat, long) in zip_lat_lon.items():
    CC_realtor_merged_cleaned_adjusted.loc[CC_realtor_merged_cleaned_adjusted['ZIP'] == zip_code, 'lat'] = lat
    CC_realtor_merged_cleaned_adjusted.loc[CC_realtor_merged_cleaned_adjusted['ZIP'] == zip_code, 'long'] = long


```

```python
# write that back to a csv
CC_realtor_merged_cleaned_adjusted.to_csv('../Data/CC_realtor_merged_cleaned_adjusted.csv', index=False)
```

```python
# get nulls of all columns
print(CC_realtor_merged_cleaned_adjusted[CC_realtor_merged_cleaned_adjusted.isnull().any(axis=1)])
```

```python
# remove ZIP, lat, long, price_multiplier, days_on_market_multiplier, adjusted_sold_price, adjusted_dom, SOLDPRICE
CC_realtor_merged_cleaned_adjusted = CC_realtor_merged_cleaned_adjusted.drop(columns=['ZIP', 'lat', 'long', 'price_multiplier', 'days_on_market_multiplier', 'adjusted_sold_price', 'adjusted_dom', 'SOLDPRICE'])

# one hot encode the zip_cluster column
#CC_realtor_merged_cleaned_adjusted = pd.get_dummies(CC_realtor_merged_cleaned_adjusted, columns=['zip_cluster'])

# set the zip_cluster column to be categorical
CC_realtor_merged_cleaned_adjusted['zip_cluster'] = CC_realtor_merged_cleaned_adjusted['zip_cluster'].astype('category')

# get head 
CC_realtor_merged_cleaned_adjusted.head()

```

```python
# check for nulls
CC_realtor_merged_cleaned_adjusted.isnull().sum()

# print the null rows
null_columns = CC_realtor_merged_cleaned_adjusted.columns[CC_realtor_merged_cleaned_adjusted.isnull().any()]
print(CC_realtor_merged_cleaned_adjusted[CC_realtor_merged_cleaned_adjusted.isnull().any(axis=1)])

```

```python
# make DOM the first column, then save to csv with header
# there are a ton of zip_cluster columns, so we can't list them out when indicating the order of the columns otherwise
CC_realtor_merged_cleaned_adjusted = pd.concat([CC_realtor_merged_cleaned_adjusted['DOM'], CC_realtor_merged_cleaned_adjusted.drop(columns=['DOM'])], axis=1)

# change the zip_cluster column to be categorical
CC_realtor_merged_cleaned_adjusted['zip_cluster'] = CC_realtor_merged_cleaned_adjusted['zip_cluster'].astype('category')

# print the column type of the zip_cluster column
print(CC_realtor_merged_cleaned_adjusted['zip_cluster'].dtype)

# save to csv with header
CC_realtor_merged_cleaned_adjusted.to_csv('../Data/baseline_model/CC_baseline_header.csv', index=False)

# save to csv without header
CC_realtor_merged_cleaned_adjusted.to_csv('../Data/baseline_model/CC_baseline_no_header.csv', index=False, header=False)

```

## Make a json with a mapping from the zip code to zip code cluster

```python
# get CC_realtor_merged_cleaned_adjusted.csv
CC_realtor_merged_cleaned_adjusted = pd.read_csv('../Data/CC_realtor_merged_cleaned_adjusted.csv')

# head
CC_realtor_merged_cleaned_adjusted.head()
```

```python
# df with zip code and zip_cluster
zip_cluster_df = CC_realtor_merged_cleaned_adjusted[['ZIP', 'zip_cluster']].drop_duplicates()

# save to json
zip_cluster_df.to_json('../Data/baseline_model/zip_cluster_mapping.json', orient='records')

```

## Make JSON for the min and max price per zip cluster

```python
# make new df with zip_cluster and min and max price
min_max_price_df = CC_realtor_merged_cleaned_adjusted.groupby('zip_cluster')['SOLDPRICE'].agg(['min', 'max']).reset_index()

# head
min_max_price_df.head()

# save to json
min_max_price_df.to_json('../Data/baseline_model/min_max_price_per_zip_cluster.json', orient='records')
```
